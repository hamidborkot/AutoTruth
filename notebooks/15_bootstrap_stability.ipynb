# Notebook 15 — Bootstrap Regime Stability

Bootstraps three-regime CDD classification across 18 TruthfulQA domains.
1000 iterations × 80% resample. Reports per-domain stability and boundary zones.

**Key result:** Core-domain stability = **94.4%**. 9/18 domains ≥85% stable.

```python
import json, pandas as pd

# Load pre-computed results
with open('../results/bootstrap_stability.json') as f:
    data = json.load(f)

df = pd.DataFrame(data['domains'])
df['stability_pct'] = (df['stability'] * 100).round(1)
df = df.sort_values('stability', ascending=False)
print(df[['domain','regime','mean_dist','stability_pct','zone']].to_string(index=False))
print(f"\nOverall mean stability: {data['summary']['overall_mean_stability']*100:.1f}%")
print(f"Core domains mean:      {data['summary']['core_domains_mean_stability']*100:.1f}%")
print(f"Stable (>=85%):         {data['summary']['domains_stable']}/{data['summary']['total_domains']}")
```
