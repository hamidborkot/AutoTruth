# CDD Notebook 13 — Stage 4: Domain-Conditional Analysis

This notebook computes domain-local causal directions and compares them to the global causal direction for each TruthfulQA category.

Expected headline result: the global Pearson correlation is weak overall because multiple mechanistic regimes are mixed.

```python
import pandas as pd, numpy as np
from scipy.stats import pearsonr

df_domain = pd.read_csv('results/autotruth_domain_regimes.csv')
r, p = pearsonr(df_domain['mean_ie'], df_domain['domain_CPA'])
print(df_domain[['category', 'mean_ie', 'domain_CPA', 'regime']].to_string(index=False))
print()
print(f'Global Pearson r = {r:.4f}, p = {p:.4f}')
```
