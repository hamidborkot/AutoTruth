# Notebook 17 — Cross-Dataset Validation: CREAK

Validates CPA on CREAK entity-centric claims (amydeng2000/CREAK).
200 same-entity contrastive pairs from 2096 Wikipedia entities.
Model: Pythia-1.4B-deduped.

**Key result:** CREAK CPA = **0.0148** — strongest near-zero result. H1 confirmed cross-dataset.

```python
import json

# Load pre-computed results
with open('../results/creak_validation.json') as f:
    data = json.load(f)

print('=== Cross-Dataset Validation: CREAK ===')
print(f"Dataset:        {data['dataset']}")
print(f"Model:          {data['model']}")
print(f"N pairs:        {data['results']['n_pairs']}")
print(f"Mean CPA:       {data['results']['mean_cpa']:.4f}")
print(f"Probe Accuracy: {data['results']['mean_probe_accuracy']:.4f}")
print(f"H1 Confirmed:   {data['results']['h1_confirmed']}")
print()
print('Cross-dataset comparison:')
for k,v in data['cross_dataset_comparison'].items():
    print(f'  {k:35s}: CPA={v:.4f}')
```
