# CDD Notebook 14 — Stage 5: Three-Regime Discovery

This notebook finalises the regime analysis, computes within-regime correlations, and prints the final three-regime summary used in the paper.

Expected headline result: Mann-Whitney p = 0.0103 and Context-Driven r = -0.781 (p = 0.038).

```python
import pandas as pd, numpy as np
from scipy.stats import mannwhitneyu, pearsonr

df = pd.read_csv('results/autotruth_domain_regimes.csv')

for regime in ['Direction-Driven', 'Context-Driven', 'Circuit Failure']:
    sub = df[df['regime'] == regime]
    print(f'{regime} (n={len(sub)}):')
    for _, row in sub.iterrows():
        print(f"  {row['category']:<30} IE={row['mean_ie']:+.3f}  CPA={row['domain_CPA']:.3f}")
    print()

dd = df[df['regime'] == 'Direction-Driven']['domain_CPA'].values
cf = df[df['regime'] == 'Circuit Failure']['domain_CPA'].values
ctx = df[df['regime'] == 'Context-Driven']

u, p_mw = mannwhitneyu(dd, cf, alternative='greater')
r_ctx, p_ctx = pearsonr(ctx['domain_CPA'], ctx['mean_ie'])

print(f'Mann-Whitney U p = {p_mw:.4f}')
print(f'Context-Driven r = {r_ctx:.3f}, p = {p_ctx:.3f}')
print(f'Direction-Driven mean CPA: {dd.mean():.3f} | mean IE: {df[df["regime"] == "Direction-Driven"]["mean_ie"].mean():.3f}')
print(f'Context-Driven mean CPA:   {ctx["domain_CPA"].mean():.3f} | mean IE: {ctx["mean_ie"].mean():.3f}')
print(f'Circuit Failure mean CPA:  {cf.mean():.3f} | mean IE: {df[df["regime"] == "Circuit Failure"]["mean_ie"].mean():.3f}')
```
